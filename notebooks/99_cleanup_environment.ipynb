#### 99 — Environment Cleanup

This notebook resets the project environment by:

- Dropping all Delta tables used in this project
- (Optional) Deleting Vector Search index

```python
# Drop all project tables

tables_to_drop = [
    RAW_DOCS_TABLE,
    CHUNKS_TABLE,
    EMB_TABLE,
    RAG_LOG_TABLE,
    RAG_EVAL_TABLE,
]

for t in tables_to_drop:
    try:
        print(f"Dropping table: {t}")
        spark.sql(f"DROP TABLE IF EXISTS {t}")
    except Exception as e:
        print(f"Failed to drop {t}: {e}")
```
